```python
import pandas as pd

try:
    df = pd.read_csv('Nan mudhalavn project ai disease 11.csv')
    display(df.head())
except FileNotFoundError:
    print("Error: File not found. Please check the file path.")
    df = None # Initialize df to None in case of error
except pd.errors.ParserError:
    print("Error: Unable to parse the file. Please check the file format.")
    df = None
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    df = None
    # Check the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)

# Examine data types
print("\nData Types:\n", df.dtypes)

# Generate descriptive statistics for numerical columns
print("\nDescriptive Statistics:\n", df.describe())

# Identify unique values and frequencies for categorical columns
print("\nUnique values and frequencies for 'Gender':\n", df['Gender'].value_counts())
print("\nUnique values and frequencies for 'DiseasePrediction':\n", df['DiseasePrediction'].value_counts())

# Check for missing values
print("\nMissing values per column:\n", df.isnull().sum())

# Examine the first few rows
print("\nFirst few rows:\n", df.head())

# Examine the last few rows
print("\nLast few rows:\n", df.tail())
import matplotlib.pyplot as plt
import seaborn as sns

# Analyze numerical features
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.histplot(df['Age'], kde=True)
plt.title('Distribution of Age')

plt.subplot(1, 3, 2)
sns.histplot(df['BloodPressure'], kde=True)
plt.title('Distribution of Blood Pressure')

plt.subplot(1, 3, 3)
sns.histplot(df['HeartRate'], kde=True)
plt.title('Distribution of Heart Rate')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.boxplot(x='DiseasePrediction', y='Age', data=df)
plt.title('Age vs. Disease Prediction')

plt.subplot(1, 3, 2)
sns.boxplot(x='DiseasePrediction', y='BloodPressure', data=df)
plt.title('Blood Pressure vs. Disease Prediction')

plt.subplot(1, 3, 3)
sns.boxplot(x='DiseasePrediction', y='HeartRate', data=df)
plt.title('Heart Rate vs. Disease Prediction')
plt.tight_layout()
plt.show()

# Investigate relationship between Gender and DiseasePrediction
gender_disease = df.groupby('Gender')['DiseasePrediction'].value_counts(normalize=True).unstack()
gender_disease.plot(kind='bar', stacked=True, figsize=(8,6))
plt.title("Disease Prediction by Gender")
plt.ylabel("Proportion")
plt.show()


# Explore relationships between pairs of numerical features
plt.figure(figsize=(10, 8))
sns.pairplot(df[['Age', 'BloodPressure', 'HeartRate', 'DiseasePrediction']], hue='DiseasePrediction')
plt.show()
correlation_matrix = df[['Age', 'BloodPressure', 'HeartRate', 'DiseasePrediction']].corr()
display(correlation_matrix)
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms with KDE
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.histplot(df['Age'], kde=True, color='skyblue')
plt.title('Distribution of Age')

plt.subplot(1, 3, 2)
sns.histplot(df['BloodPressure'], kde=True, color='salmon')
plt.title('Distribution of Blood Pressure')

plt.subplot(1, 3, 3)
sns.histplot(df['HeartRate'], kde=True, color='lightgreen')
plt.title('Distribution of Heart Rate')
plt.tight_layout()
plt.show()

# Box plots
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.boxplot(x='DiseasePrediction', y='Age', data=df, palette="pastel")
plt.title('Age vs. Disease Prediction')

plt.subplot(1, 3, 2)
sns.boxplot(x='DiseasePrediction', y='BloodPressure', data=df, palette="pastel")
plt.title('Blood Pressure vs. Disease Prediction')

plt.subplot(1, 3, 3)
sns.boxplot(x='DiseasePrediction', y='HeartRate', data=df, palette="pastel")
plt.title('Heart Rate vs. Disease Prediction')
plt.tight_layout()
plt.show()

# Stacked bar chart
gender_disease = df.groupby('Gender')['DiseasePrediction'].value_counts(normalize=True).unstack()
gender_disease.plot(kind='bar', stacked=True, figsize=(8,6), color=['skyblue', 'salmon'])
plt.title("Disease Prediction by Gender")
plt.ylabel("Proportion")
plt.show()

# Pair plot
plt.figure(figsize=(10, 8))
sns.pairplot(df[['Age', 'BloodPressure', 'HeartRate', 'DiseasePrediction']], hue='DiseasePrediction', palette="Set1")
plt.show()

# Correlation heatmap
correlation_matrix = df[['Age', 'BloodPressure', 'HeartRate', 'DiseasePrediction']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms with KDE
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.histplot(df['Age'], kde=True, color='skyblue')
plt.title('Distribution of Age')

plt.subplot(1, 3, 2)
sns.histplot(df['BloodPressure'], kde=True, color='salmon')
plt.title('Distribution of Blood Pressure')

plt.subplot(1, 3, 3)
sns.histplot(df['HeartRate'], kde=True, color='lightgreen')
plt.title('Distribution of Heart Rate')
plt.tight_layout()
plt.show()

# Box plots - Corrected to remove warnings
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.boxplot(x='DiseasePrediction', y='Age', data=df, palette="pastel", showfliers=False) #showfliers=False to remove outliers
plt.title('Age vs. Disease Prediction')

plt.subplot(1, 3, 2)
sns.boxplot(x='DiseasePrediction', y='BloodPressure', data=df, palette="pastel", showfliers=False)
plt.title('Blood Pressure vs. Disease Prediction')

plt.subplot(1, 3, 3)
sns.boxplot(x='DiseasePrediction', y='HeartRate', data=df, palette="pastel", showfliers=False)
plt.title('Heart Rate vs. Disease Prediction')
plt.tight_layout()
plt.show()

# Stacked bar chart
gender_disease = df.groupby('Gender')['DiseasePrediction'].value_counts(normalize=True).unstack()
gender_disease.plot(kind='bar', stacked=True, figsize=(8,6), color=['skyblue', 'salmon'])
plt.title("Disease Prediction by Gender")
plt.ylabel("Proportion")
plt.show()

# Pair plot
sns.pairplot(df[['Age', 'BloodPressure', 'HeartRate', 'DiseasePrediction']], hue='DiseasePrediction', palette="Set1")
plt.show()

# Correlation heatmap
correlation_matrix = df[['Age', 'BloodPressure', 'HeartRate', 'DiseasePrediction']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms with KDE
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.histplot(df['Age'], kde=True, color='skyblue')
plt.title('Distribution of Age')

plt.subplot(1, 3, 2)
sns.histplot(df['BloodPressure'], kde=True, color='salmon')
plt.title('Distribution of Blood Pressure')

plt.subplot(1, 3, 3)
sns.histplot(df['HeartRate'], kde=True, color='lightgreen')
plt.title('Distribution of Heart Rate')
plt.tight_layout()
plt.show()

# Box plots - Corrected to remove warnings and explicitly set hue
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
sns.boxplot(x='DiseasePrediction', y='Age', hue='DiseasePrediction', data=df, palette="pastel", showfliers=False, legend=False)
plt.title('Age vs. Disease Prediction')

plt.subplot(1, 3, 2)
sns.boxplot(x='DiseasePrediction', y='BloodPressure', hue='DiseasePrediction', data=df, palette="pastel", showfliers=False, legend=False)
plt.title('Blood Pressure vs. Disease Prediction')

plt.subplot(1, 3, 3)
sns.boxplot(x='DiseasePrediction', y='HeartRate', hue='DiseasePrediction', data=df, palette="pastel", showfliers=False, legend=False)
plt.title('Heart Rate vs. Disease Prediction')
plt.tight_layout()
plt.show()

# Stacked bar chart
gender_disease = df.groupby('Gender')['DiseasePrediction'].value_counts(normalize=True).unstack()
gender_disease.plot(kind='bar', stacked=True, figsize=(8,6), color=['skyblue', 'salmon'])
plt.title("Disease Prediction by Gender")
plt.ylabel("Proportion")
plt.show()

# Pair plot
sns.pairplot(df[['Age', 'BloodPressure', 'HeartRate', 'DiseasePrediction']], hue='DiseasePrediction', palette="Set1")
plt.show()

# Correlation heatmap
correlation_matrix = df[['Age', 'BloodPressure', 'HeartRate', 'DiseasePrediction']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```
